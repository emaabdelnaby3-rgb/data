### First prepare your data first (nulls , missings and so on ..)

```python
import pandas as pd

file = pd.read_csv(
    "C:\\Users\\Admin\\Downloads\\google-play-store\\google-play-store.csv"
)
file.head()
file.info()
# b show el nulls
file.isna().sum()
# hadrop reviews ratings score cuz there is a 11 row null w subset bt5aleny abos 3la el rows el 3awza ashelha bs
file = file.dropna(subset=["reviews", "ratings", "score"])
file = file.dropna(subset=["genre", "genre_id"])
file.isna().sum()
```

```python
# bashof eh dol
file[file["released"].isna()].head()
# hamlahom bl mediam cuz i can't drop them
# before drop it ht change el type from obj to date
file["released"] = pd.to_datetime(file["released"])
# ha fill them with median
file["released"].fillna(file["released"].median(), inplace=True)
file.info()
file.isna().sum()
```
